```python
from osgeo import ogr, osr

# set up the shapefile driver
driver = ogr.GetDriverByName("ESRI Shapefile")

# input SpatialReference
inSpatialRef = osr.SpatialReference()
inSpatialRef.ImportFromEPSG(4269)  # replace this with the EPSG code of your shapefile if it's not WGS84

# output SpatialReference
outSpatialRef = osr.SpatialReference()
outSpatialRef.ImportFromEPSG(4326)

# create the CoordinateTransformation
coordTrans = osr.CoordinateTransformation(inSpatialRef, outSpatialRef)

# get the input layer
inDataSet = driver.Open(r'C:\Users\eobasi\Downloads\pr33bndg.shp')
inLayer = inDataSet.GetLayer()

# create the output KML file
outDriver = ogr.GetDriverByName('KML')
outDataSet = outDriver.CreateDataSource('output.kml')
outLayer = outDataSet.CreateLayer('output', geom_type=ogr.wkbPolygon)

# add features to the ouput Layer
for inFeature in inLayer:
    geom = inFeature.GetGeometryRef()
    geom.Transform(coordTrans)
    outFeature = ogr.Feature(outLayer.GetLayerDefn())
    outFeature.SetGeometry(geom)
    outLayer.CreateFeature(outFeature)
    outFeature = None

# Save and close DataSources
inDataSet = None
outDataSet = None
```
